```python
from scipy.stats import zscore
z_score=abs(zscore(df))
print(df.shape)
df_new=df.loc[(z_score<3).all(axis=1)]
print(df_new.shape)
```

```python
df_new.skew()# checking skewness
```

```python
df_new.shape
```

```python
from scipy.stats import boxcox
# 0= log transform
# .5 = square root transform
df_new['Standard Error']=boxcox(df_new['Standard Error'],0)
df_new['Generosity']=boxcox(df_new['Generosity'],0)
```
